**Persuasive words detector function**

Parameters:

-text: String where persuasive words will be searched

-persuasiveWordsDict: Dictionary with the persuasive words

Return: vector list with 0 or 1 (not found or found) in the corresponding position of persuasive words dictionary

```python
from re import search
def Persuasive_Words_Detector(text, persuasiveWordsDict):
  resultVector = []
  for key, value in persuasiveWordsDict.items():
    ps = PorterStemmer()
    stem = ps.stem(key)
    #se crea expresion regular para identificar exactamente la palabra de persuacion que se busca, evitando falso positivo cuando forma parte de otra palabra
    regEx = "(^"+ key[0] +"|[^a-z]"+ key[0] +")"+ key[1:-1] +"("+key[-1]+"[^a-z]|"+key[-1]+"$)"
    #print (regEx)
    if search(regEx, text.lower()):
      resultVector.append("1")
    else:
      resultVector.append("0")
  return resultVector
```

Testing the previous function

```python
fileWords = open('/content/gdrive/MyDrive/Colabs/Data/phishing_esp/wordsVector_legit.csv', 'w')
header = ''
for num, key in enumerate(persuasive_words_dict.keys()):
  if num != 0:
    header += ';'
  header += key
fileWords.write(header + '\n')
print(header)

vectors = []
with open('/content/gdrive/MyDrive/Colabs/Data/phishing_esp/dataset_legit.txt', encoding = 'unicode_escape') as f:
  for line in f:
    txt = ""
    # Lemmatization with stopwords removal
    txt = line.split()

    ps = PorterStemmer()
    stem = ""

    for word in txt:
      words = word_tokenize(word)
      for w in words:
        stem += ps.stem(w) + " "

    vectorLst = Persuasive_Words_Detector(stem, persuasive_words_dict)
    lineTxt = ''
    for pos,key in enumerate(persuasive_words_dict.keys()):
      if pos != 0:
        lineTxt += ';'
      lineTxt += vectorLst[pos]
    vectors.append(lineTxt + '\n')
    print(lineTxt)
fileWords.writelines(vectors)
fileWords.close()
```
